```python
#specifying the column names that I want to use
new_cols_dict ={
    'timestamp':'time',
    'cnt':'count',
    't1':'temp_real_C',
    't2':'temp_feels_like_C',
    'hum':'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

#renaming the columns to the specified column names
bikes_data.rename(new_cols_dict, axis=1, inplace=True)
```

```python
#changing the humidity values to percentage (i.e value between 0 and 1)
bikes_data.humidity_percent = bikes_data.humidity_percent / 100
```

```python
#creating a season dictionary so that we can map the integers 0-3 to the actual written values
season_dict ={
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter'
}

#creating a weather dictionary so that we can map the integers to the actual values
weather_dict ={
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}
```

```python
#changing the seasons column data type to string
bikes_data.season = bikes_data.season.astype('str')
#mapping the values 0-3 to the actual written seasons
bikes_data.season = bikes_data.season.map(season_dict)

#changing the weather column data type to string
bikes_data.weather = bikes_data.weather.astype('str')
#mapping the values to the actual written weathers
bikes_data.weather = bikes_data.weather.map(weather_dict)
```

```python
#checking the dataframe to see if the mappings have worked
bikes_data.head()
```
